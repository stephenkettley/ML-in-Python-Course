### Import Relevant Libraries

```python
raw_data = pd.read_csv('car-sales.csv')
raw_data.head()
```

### Preprocessing

#### Relaxing the Assumptions

```python
log_price = np.log(data_cleaned['Price'])
data_cleaned['log Price'] = log_price
data_cleaned
```

```python
f, (ax1,ax2,ax3) = plt.subplots(1, 3, sharey=True, figsize=(15,3))
ax1.scatter(data_cleaned['Year'], data_cleaned['log Price'])
ax1.set_title('log Price and Year')
ax2.scatter(data_cleaned['EngineV'], data_cleaned['log Price'])
ax2.set_title('log Price and EngineV')
ax3.scatter(data_cleaned['Mileage'], data_cleaned['log Price'])
ax3.set_title('log Price and Mileage')

plt.show() #by using the log price we will obtain linear patterns
```

```python
data_cleaned = data_cleaned.drop(['Price'], axis=1)
```

#### Multicollinearity

```python
data_cleaned.columns.values
```

```python
from statsmodels.stats.outliers_influence import variance_inflation_factor
variables = data_cleaned[['Mileage','Year','EngineV']]
vif = pd.DataFrame()
vif['VIF'] = [variance_inflation_factor(variables.values, i) for i in range(variables.shape[1])]
vif['features'] = variables.columns
vif
```

```python
data_no_multicollinearity = data_cleaned.drop(['Year'], axis=1)
```

### Create Dummy Variables

```python
data_with_dummies = pd.get_dummies(data_no_multicollinearity, drop_first=True)
```

```python
data_with_dummies.head()
```

### Rearrange Columns


```python
data_with_dummies.columns.values
```

```python
cols = ['log Price', 'Mileage', 'EngineV', 'Brand_BMW',
       'Brand_Mercedes-Benz', 'Brand_Mitsubishi', 'Brand_Renault',
       'Brand_Toyota', 'Brand_Volkswagen', 'Body_hatch', 'Body_other',
       'Body_sedan', 'Body_vagon', 'Body_van', 'Engine Type_Gas',
       'Engine Type_Other', 'Engine Type_Petrol', 'Registration_yes']
```

```python
data_preprocessed = data_with_dummies[cols]
data_preprocessed.head()
```

### Linear Regression Model

#### Declare the Inputs and Targets

```python
targets = data_preprocessed['log Price']
inputs = data_preprocessed.drop(['log Price'], axis=1)
```

#### Scale the Data

```python
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()
scaler.fit(inputs)
```

```python
inputs_scaled = scaler.transform(inputs)
```

#### Train Test Split

```python
from sklearn.model_selection import train_test_split
```

```python
x_train, x_test, y_train, y_test = train_test_split(inputs_scaled, targets, test_size=0.2, random_state=365)
```

#### Create the Regression

```python
reg = LinearRegression()
reg.fit(x_train, y_train)
```

```python
y_hat = reg.predict(x_train)
```

```python
plt.scatter(y_train, y_hat) #scatter of the targets and the predictions
plt.xlabel('Targets (y train)')
plt.ylabel('Predictions (y hat)')
plt.xlim(6,13)
plt.ylim(6,13)
plt.show()
```

```python
sns.displot(y_train - y_hat)
plt.title('Residuals PDF')
plt.show()
```

```python
reg.score(x_train, y_train)
```

#### Finding the Weights and Bias

##### Bias

```python
reg.intercept_
```

##### Weights

```python
reg.coef_
```

```python
reg_summary = pd.DataFrame(inputs.columns.values, columns=['Features'])
reg_summary['Weights'] = reg.coef_
reg_summary
```

### Testing

```python
y_hat_test = reg.predict(x_test)
```

```python
plt.scatter(y_test, y_hat_test, alpha=0.2) 
plt.xlabel('Targets (y test)')
plt.ylabel('Predictions (y hat test)')
plt.xlim(6,13)
plt.ylim(6,13)
plt.show()
```

```python
df_pf = pd.DataFrame(np.exp(y_hat_test), columns=['Predictions'])
df_pf.head()
```

```python
y_test = y_test.reset_index(drop=True)
df_pf['Targets'] = np.exp(y_test)
df_pf.head()
```

```python
df_pf['Residual'] = df_pf['Predictions'] - df_pf['Targets']
df_pf.head()
```

```python
df_pf['Difference%'] = np.absolute(df_pf['Residual']/df_pf['Targets']*100)
df_pf.head()
```
